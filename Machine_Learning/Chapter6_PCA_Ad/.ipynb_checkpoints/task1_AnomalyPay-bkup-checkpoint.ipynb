```python
#load the data
import pandas as pd
import numpy as np
data = pd.read_csv('ad_data3.csv')
data.head()
```

```python
#visualize the data
%matplotlib inline
from matplotlib import pyplot as plt

fig1 = plt.figure(figsize=(10,6))
plt.scatter(data.loc[:,'x1'],data.loc[:,'x2'],marker='x')
plt.title('data')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
#define x1 and x2
x1 = data.loc[:,'x1']
x2 = data.loc[:,'x2']
```

```python
# coding:utf-8
import matplotlib as mlp
font2 = {'family' : 'SimHei',
'weight' : 'normal',
'size'   : 20,
}
mlp.rcParams['font.family'] = 'SimHei'
mlp.rcParams['axes.unicode_minus'] = False
fig2 = plt.figure(figsize=(20,5))
plt.subplot(121)
plt.hist(x1,bins=100)
plt.title('$x_1$ 数据分布统计',font2)
plt.xlabel('$x_1$',font2)
plt.ylabel('出现次数',font2)
plt.subplot(122)
plt.hist(x2,bins=100)
plt.title('x2 distribution')
plt.xlabel('x2')
plt.ylabel('counts')
plt.show()
```

```python
#calculate the mean and sigma of x1 and x2
x1_mean = x1.mean()
x1_sigma = x1.std()
x2_mean = x2.mean()
x2_sigma = x2.std()
print(x1_mean,x1_sigma,x2_mean,x2_sigma)
```

```python
#calculate the gaussian distribution p(x)
from scipy.stats import norm
x1_range = np.linspace(0,4,300)
x1_normal = norm.pdf(x1_range,x1_mean,x1_sigma)
x2_range = np.linspace(0,10,300)
x2_normal = norm.pdf(x2_range,x2_mean,x2_sigma)
```

```python
#visualize the p(x)
fig2 = plt.figure(figsize=(20,5))
plt.subplot(121)
plt.plot(x1_range,x1_normal)
plt.title('normal p(x1)')
plt.subplot(122)
plt.plot(x2_range,x2_normal)
plt.title('normal p(x2)')
plt.show()
```

```python
import math
#设置范围
x_min, x_max = 0, 4
y_min, y_max = 0, 10
h = 0.05
#生成矩阵数据
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
print(xx.shape,yy.shape)

#展开矩阵数据
x_range = np.c_[xx.ravel(), yy.ravel()]
x1 = np.c_[xx.ravel()]
x2 = np.c_[yy.ravel()]
x_range_df = pd.DataFrame(x_range)
#x_range_df.to_csv('data.csv')
#高斯分布参数
u1 = x1_mean
u2 = x2_mean
sigma1 = x1_sigma
sigma2 = x2_sigma

#计算高斯分布概率
p1 = 1/sigma1/math.sqrt(2*math.pi)*np.exp(-np.power((x1-u1),2)/2/math.pow(sigma1,2))
p2 = 1/sigma2/math.sqrt(2*math.pi)*np.exp(-np.power((x2-u2),2)/2/math.pow(sigma2,2))
p = np.multiply(p1,p2)
#对概率密度维度转化
p_2d = p.reshape(xx.shape[0],xx.shape[1])
print(p_2d)
#图形可视化
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
fig = plt.figure()
axes3d = Axes3D(fig)
 
axes3d.plot_surface(xx,yy,p_2d,cmap=cm.rainbow)
axes3d.set_xlabel(xlabel= r'$x_1$')
axes3d.set_ylabel(ylabel= r'$x_2$')
axes3d.set_zlabel(zlabel= r'$p(x_1,x_2)$')
```

```python
#establish the model and predict
from sklearn.covariance import EllipticEnvelope
ad_model = EllipticEnvelope(contamination=0.03)
ad_model.fit(data)
```

```python
#make prediction
y_predict = ad_model.predict(data)
print(pd.value_counts(y_predict))
```

```python
#visualize the result
fig4 = plt.figure(figsize=(10,6))
orginal_data=plt.scatter(data.loc[:,'x1'],data.loc[:,'x2'],marker='x')
anomaly_data=plt.scatter(data.loc[:,'x1'][y_predict==-1],data.loc[:,'x2'][y_predict==-1],marker='o',facecolor='none',edgecolor='red',s=150)

plt.title('自动寻找异常数据',font2)
plt.xlabel('$x_1$',font2)
plt.ylabel('$x_2$',font2)
plt.legend((orginal_data,anomaly_data),('原数据','检测异常点'))
plt.axis([0,5,0,12])
plt.show()
```

```python
ad_model = EllipticEnvelope(contamination=0.2)
ad_model.fit(data)
y_predict = ad_model.predict(data)
#visualize the result
fig5 = plt.figure(figsize=(20,10))
orginal_data=plt.scatter(data.loc[:,'x1'],data.loc[:,'x2'],marker='x')
anomaly_data=plt.scatter(data.loc[:,'x1'][y_predict==-1],data.loc[:,'x2'][y_predict==-1],marker='o',facecolor='none',edgecolor='red',s=150)

plt.title('anomaly detection result')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend((orginal_data,anomaly_data),('original_data','anomaly_data'))
plt.show()
```

```python
#生成gif动图
import PIL.Image as Image
def get_gif(pic_dir,n,t=1):
    imgs = []
    for i in range(1,n):
        pic_name = '{}/{}.png'.format(pic_dir,i)
        temp = Image.open(pic_name)
        imgs.append(temp)
    save_name = '{}.gif'.format(pic_dir)
    imgs[0].save(save_name, save_all=True, append_images=imgs, duration=3)
    return save_name


pic_dir = 'ad_output'
save_name = get_gif(pic_dir,20,1)
print('制作完成。文件为:{}'.format(save_name))
```
