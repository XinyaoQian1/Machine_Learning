```python
import math
#设置范围
x_min, x_max = 1, 5
y_min, y_max = 3, 5
h = 0.05
#生成矩阵数据
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
print(xx.shape,yy.shape)

#展开矩阵数据
x_range = np.c_[xx.ravel(), yy.ravel()]
x1 = np.c_[xx.ravel()]
x2 = np.c_[yy.ravel()]
#高斯分布参数
u1 = 3
u2 = 4
sigma1 = 0.5
sigma2 = 0.14

#计算高斯分布概率
p1 = 1/sigma1/math.sqrt(2*math.pi)*np.exp(-np.power((x1-u1),2)/2/math.pow(sigma1,2))
p2 = 1/sigma2/math.sqrt(2*math.pi)*np.exp(-np.power((x2-u2),2)/2/math.pow(sigma2,2))
p = np.multiply(p1,p2)
#对概率密度维度转化
p_2d = p.reshape(40,80)
```

```python
#图形可视化
from matplotlib import cm
fig = plt.figure()
axes3d = Axes3D(fig)
 
axes3d.plot_surface(xx,yy,p_2d,cmap=cm.rainbow)
```
