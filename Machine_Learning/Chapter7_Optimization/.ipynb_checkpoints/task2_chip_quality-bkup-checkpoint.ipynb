```python
import math
x1 = data.loc[:,'x1'][y==0]
x2 = data.loc[:,'x2'][y==0]
#calculate the mean and sigma of x1 and x2
x1_mean = x1.mean()
x1_sigma = x1.std()
x2_mean = x2.mean()
x2_sigma = x2.std()
print(x1_mean,x1_sigma,x2_mean,x2_sigma)
#计算高斯分布概率
u1 = x1_mean
u2 = x2_mean
sigma1 = x1_sigma
sigma2 = x2_sigma
p1 = 1/sigma1/math.sqrt(2*math.pi)*np.exp(-np.power((x1-u1),2)/2/math.pow(sigma1,2))
p2 = 1/sigma2/math.sqrt(2*math.pi)*np.exp(-np.power((x2-u2),2)/2/math.pow(sigma2,2))
p = np.multiply(p1,p2)
print(p)
print('max p:',max(p),'\n','min p:',min(p),'\n','max/min:',max(p)/min(p))
```

```python
# data = pd.read_csv('data_class_processed.csv')
# data.head()
#define X and y
X = data.drop(['y'],axis=1)
y = data.loc[:,'y']
X = np.array(X)
```

```python
#主成分分析
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
X_norm = StandardScaler().fit_transform(X)
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_norm)
var_ratio = pca.explained_variance_ratio_
print(var_ratio)
fig4 = plt.figure(figsize=(5,5))
plt.bar([1,2],var_ratio)
plt.xticks([1,2],['PC1','PC2'])
plt.show()
```

```python
# 数据分离: random_state=4,test_size=0.4
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1,test_size=0.4)
print(X_train.shape,X_test.shape,X.shape)
print(X_test)
```

```python
#knn模型
from sklearn.neighbors import KNeighborsClassifier
knn_3 = KNeighborsClassifier(n_neighbors=3)
knn_3.fit(X_train,y_train)
y_train_predict = knn_3.predict(X_train)
y_test_predict = knn_3.predict(X_test)

#计算准确率
from sklearn.metrics import accuracy_score
accuracy_train = accuracy_score(y_train,y_train_predict)
accuracy_test = accuracy_score(y_test,y_test_predict)
print("trianing accuracy:",accuracy_train)
print('testing accuracy:',accuracy_test)

```

```python
#计算混淆矩阵
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_test_predict)
print(cm)
```

![confusion_matrix](images/06_confusion_matrix.png)

```python
#获取混淆矩阵各元素
TP = cm[1,1]
TN = cm[0,0]
FP = cm[0,1]
FN = cm[1,0]
print(TP,TN,FP,FN)
```

**准确率:** 整体样本中，预测正确样本数的比例
- Accuracy = (TP + TN)/(TP + TN + FP + FN)

```python
accuracy = (TP + TN)/(TP + TN + FP + FN)
print(accuracy)
```

**灵敏度（召回率）:** 正样本中，预测正确的比例
- Sensitivity = Recall = TP/(TP + FN)

```python
recall = TP/(TP + FN)
print(recall)
```

**特异度:** 负样本中，预测正确的比例
- Specificity = TN/(TN + FP)

```python
specificity = TN/(TN + FP)
print(specificity)
```

**精确率:** 预测结果为正的样本中，预测正确的比例
- Precision = TP/(TP + FP)

```python
precision = TP/(TP + FP)
print(precision)
```

**F1分数:** 综合Precision和Recall的一个判断指标
- F1 Score = 2*Precision X Recall/(Precision + Recall)

```python
f1 = 2*precision*recall/(precision+recall)
print(f1)
```

```python
#try different k and calcualte the accuracy for each
n = [i for i in range(1,21)]
accuracy_train = []
accuracy_test = []
for i in n:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_train_predict = knn.predict(X_train)
    y_test_predict = knn.predict(X_test)
    accuracy_train_i = accuracy_score(y_train,y_train_predict)
    accuracy_test_i = accuracy_score(y_test,y_test_predict)
    accuracy_train.append(accuracy_train_i)
    accuracy_test.append(accuracy_test_i)
print(accuracy_train,accuracy_test)
```

```python
fig5 = plt.figure(figsize=(12,5))
plt.subplot(121)
plt.plot(n,accuracy_train,marker='o')
plt.title('不同K值的影响（训练数据）',font2)
plt.xlabel('K值',font2)
plt.ylabel('训练数据预测准确率',font2)

plt.subplot(122)
plt.plot(n,accuracy_test,marker='o')
plt.title('不同K值的影响（测试数据）',font2)
plt.xlabel('K值',font2)
plt.ylabel('测试数据预测准确率',font2)
plt.show()
```
