```python
# 数据加载：
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

cv_data = pd.read_csv('task1_data.csv')
cv_data.head()
```

```python
skill = cv_data.loc[:,'Skill']
#找出skill分为几类（也就是不重复的元素有多少个)
np.unique(skill)
```

```python
inputs = cv_data.drop('y',axis=1)

y = cv_data.loc[:,'y']
np.unique(y)
#维度确认
print(inputs.shape,y.shape)
```

```python
#建立决策树模型
from sklearn.tree import DecisionTreeClassifier
Dtree = DecisionTreeClassifier(criterion='entropy',min_samples_leaf= 5)
 	#criterion: (评判或作决定的)标准，准则，原则
```

```python
#模型训练
Dtree.fit(inputs,y)
```

```python
#模型预测
y_pre = Dtree.predict(inputs)
np.unique(y_pre)
```

```python
#评估预测结果
from sklearn.metrics import accuracy_score
Daccuracy = accuracy_score(y,y_pre)
print(Daccuracy)
```
