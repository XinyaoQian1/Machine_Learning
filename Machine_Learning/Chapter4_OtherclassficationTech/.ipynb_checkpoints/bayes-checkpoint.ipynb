```python
#load the data
import pandas as pd
import numpy as np
data = pd.read_csv('task2_data.csv')
data.head(10)
```

```python
#define the X and y
X = data.drop(['y'],axis=1)
y = data.loc[:,'y']
print(X.shape,y.shape)
```

```python
#establish the naive_bayes model
from sklearn.naive_bayes import CategoricalNB

model = CategoricalNB()
model.fit(X, y)
```

```python
y_predict_prob = model.predict_proba(X)
print(y_predict_prob)
```

```python
#输出预测的y
y_predict = model.predict(X)
print(y_predict)
```

```python
#计算模型准确率
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y,y_predict)
print(accuracy)
```

```python
#测试样本的预测
X_test = np.array([[2,1,1,1,1]])
y_test_proba = model.predict_proba(X_test)
print(y_test_proba)
```

```python
y_test = model.predict(X_test)
print(y_test)
```

```python
X_test = np.array([[2,1,1,1,1],[2,1,1,1,0],[2,1,1,0,0],[2,1,0,0,0],[2,0,0,0,0]])
y_test_proba = model.predict_proba(X_test)
print(y_test_proba)
```

```python
y_test = model.predict(X_test)
print(y_test)
```

```python
print(X_test)
print(y_test_proba)
print(np.array(y_test).reshape(5,1))
```

```python
#结果组合输出
test_data_result = np.concatenate((X_test, y_test_proba,np.array(y_test).reshape(5,1)), axis=1)
```

```python
test_data_result = pd.DataFrame(test_data_result)
test_data_result.head()
```

```python
test_data_result.columns=['score','school','award','gender','english','p0','p1','p2','y_predict']
test_data_result.head()
```

```python
test_data_result.to_csv('test_data_result.csv')
```
