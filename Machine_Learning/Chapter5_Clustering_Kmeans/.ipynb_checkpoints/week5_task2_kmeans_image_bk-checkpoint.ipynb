```python
#图像加载与展示
import numpy as np
import matplotlib.pyplot as plt
from skimage import io as io
img = io.imread("1.jpg")
plt.imshow(img)
```

```python
#查看数据结构及维度
print(type(img))
print(img.shape)
```

```python
print(img)
```

```python
img_width = img.shape[1]
img_height = img.shape[0]
print(img_width,img_height)
```

```python
#数据维度转化
img_data = np.reshape(img,(-1,3))
print(img_data.shape)
print(img_data)
```

```python
#模型建立与训练
from sklearn.cluster import KMeans
KM = KMeans(n_clusters=3,random_state=0)#聚类中心的个数为3
KM.fit(img_data)
```

```python
#聚类结果预测
label=KM.predict(img_data)
print(label.shape)
print(label)
```

```python
#维度度转化
label = label.reshape([img_height,img_width])   
print(label.shape)
print(label)
```

```python
#聚类类别结果转灰度值
label = 1/(label+1)
print(label.shape)
print(label)
```

```python
#图片保存到本地
io.imsave("test4.png",label)
```

```python
plt.imshow(label)
```

```python
KM = KMeans(n_clusters=8,random_state=0)#聚类中心的个数为3
KM.fit(img_data)
label=KM.predict(img_data)
label = label.reshape([img_height,img_width])    #聚类获得每个像素所属的类别
print(label.shape)
print(label)
```

```python
label = 1/(label+1)
io.imsave("test8.png",label)
plt.imshow(label)
```
