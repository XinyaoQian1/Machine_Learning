# KMeans实现数据聚类

```python
#数据加载
import pandas as pd
import numpy as np
data = pd.read_csv('task1_data1.csv')
data_result = pd.read_csv('task1_data2.csv')
data.head()
# data_result.head()
```

```python
#获取唯一一个有标签的数据点
X_labeled = data.iloc[0,:]
print(X_labeled)
print(type(X_labeled))
```

```python
#X赋值
X = data.drop(['y'],axis=1)
X.head()
```

```python
#获取用于模型评估的正确结果
y = data_result.loc[:,'y']
y.head()
```

```python
#数据可视化
from matplotlib import pyplot as plt
import matplotlib as mlp
font2 = {'family':'SimHei','weight':'normal','size':14}
fig1 = plt.figure()
plt.scatter(X.loc[:,'x1'],X.loc[:,'x2'],label='unlabeled')
plt.scatter(X_labeled['x1'],X_labeled['x2'],label='1abeled')
plt.title("原始数据分布",font2)
plt.xlabel('x1',font2)
plt.ylabel('x2',font2)
plt.legend(loc='upper left')
plt.show()
```

```python
#带标签数据可视化
fig1 = plt.figure()
label0 = plt.scatter(X.loc[:,'x1'][y==0],X.loc[:,'x2'][y==0])
label1 = plt.scatter(X.loc[:,'x1'][y==1],X.loc[:,'x2'][y==1])

plt.title("labled data")
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend((label0,label1),('label0','label1'), loc='upper left')
plt.show()
```

```python
#模型建立与训练
from sklearn.cluster import KMeans
KM = KMeans(n_clusters=2,init='random',random_state=0)
KM.fit(X)
```

```python
#查看聚类中心
centers = KM.cluster_centers_
print(centers)
```

```python
#聚类中心可视化
fig3 = plt.figure()
label0 = plt.scatter(X.loc[:,'x1'][y==0],X.loc[:,'x2'][y==0],label='label0')
label1 = plt.scatter(X.loc[:,'x1'][y==1],X.loc[:,'x2'][y==1],label='label1')

plt.title("labled data")
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(centers[:,0],centers[:,1],100,marker='x',label='centers')
# plt.legend((label0,label1),('label0','label1'), loc='upper left')
plt.legend()
plt.show()
```

```python
#结果预测
y_predict = KM.predict(X)
print(pd.value_counts(y_predict),'\n')
print(pd.value_counts(y))
```

```python
#准确率计算
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y,y_predict)
print(accuracy)
```

```python
#可视化预测结果
import matplotlib as mlp
font2 = {'family' : 'SimHei',
'weight' : 'normal',
'size'   : 20,
}
plt.figure(figsize=(16,8))
fig4 = plt.subplot(121)
label0 = plt.scatter(X.loc[:,'x1'][y_predict==0],X.loc[:,'x2'][y_predict==0])
label1 = plt.scatter(X.loc[:,'x1'][y_predict==1],X.loc[:,'x2'][y_predict==1])

plt.title("predicted data",font2)
plt.xlabel('x1',font2)
plt.ylabel('x2',font2)
plt.legend((label0,label1),('label0','label1'), loc='upper left',prop=font2)
plt.scatter(centers[:,0],centers[:,1])

fig5 = plt.subplot(122)
label0 = plt.scatter(X.loc[:,'x1'][y==0],X.loc[:,'x2'][y==0])
label1 = plt.scatter(X.loc[:,'x1'][y==1],X.loc[:,'x2'][y==1])

plt.title("labled data",font2)
plt.xlabel('x1',font2)
plt.ylabel('x2',font2)
plt.legend((label0,label1),('label0','label1'), loc='upper left',prop=font2)
plt.scatter(centers[:,0],centers[:,1])
plt.show()
```

```python
#查看并对比第一个样本点
print(X_labeled)
print(y_predict[0])
```

```python
#结果矫正(类别0矫正为1，类别1矫正为0)
y_corrected = []
for i in y_predict:
    if i==0:
        y_corrected.append(1)
    elif i==1:
        y_corrected.append(0)

print(pd.value_counts(y_corrected),'\n',pd.value_counts(y))
```

```python
print(accuracy_score(y,y_corrected))
```

```python
y_corrected = np.array(y_corrected)
print(type(y_corrected))
```

```python
plt.figure(figsize=(16,6))
fig6 = plt.subplot(121)
label0 = plt.scatter(X.loc[:,'x1'][y_corrected==0],X.loc[:,'x2'][y_corrected==0])
label1 = plt.scatter(X.loc[:,'x1'][y_corrected==1],X.loc[:,'x2'][y_corrected==1])

plt.title("KMeans矫正后预测结果（accuracy=0.99）",font2)
plt.xlabel('x1',font2)
plt.ylabel('x2',font2)
plt.legend((label0,label1),('label0','label1'), loc='upper left',prop=font2)
#plt.scatter(centers[:,0],centers[:,1])

fig7 = plt.subplot(122)
label0 = plt.scatter(X.loc[:,'x1'][y==0],X.loc[:,'x2'][y==0])
label1 = plt.scatter(X.loc[:,'x1'][y==1],X.loc[:,'x2'][y==1])

plt.title("KNN预测结果（accuracy=1）",font2)
plt.xlabel('x1',font2)
plt.ylabel('x2',font2)
plt.legend((label0,label1),('label0','label1'), loc='upper left',prop=font2)
#plt.scatter(centers[:,0],centers[:,1])
plt.show()
```

```python
#建立KNN模型及训练
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=3)
KNN.fit(X,y)
```

```python
print(pd.value_counts(y_predict_knn),pd.value_counts(y))
```

```python
#迭代一次的结果
KM = KMeans(n_clusters=2,init='random',random_state=1,n_init=1,max_iter=1)
KM.fit(X)

centers = KM.cluster_centers_
#predict based on training data
y_predict = KM.predict(X)

#visualize the data and results
fig_i = plt.figure()
label0 = plt.scatter(X.loc[:,'x1'][y_predict==0],X.loc[:,'x2'][y_predict==0])
label1 = plt.scatter(X.loc[:,'x1'][y_predict==1],X.loc[:,'x2'][y_predict==1])

plt.title("predicted data")
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend((label0,label1),('label0','label1'), loc='upper left')
plt.scatter(centers[:,0],centers[:,1])
#fig_i.savefig('{}.png'.format(i),dpi=500,bbox_inches = 'tight')
```

```python
#逐步迭代查看KMeans模型训练效果
centers = np.array([[0,0,0,0]])
for i in range(1,10):
    KM = KMeans(n_clusters=2,random_state=1,init='random',n_init=1,max_iter=i)
    KM.fit(X)

    centers_i = KM.cluster_centers_
    centers_i_temp = centers_i.reshape(1,-1)
    centers = np.concatenate((centers,centers_i_temp),axis=0)
    #predict based on training data
    y_predict = KM.predict(X)

    #visualize the data and results
    fig_i = plt.figure()
    label0 = plt.scatter(X.loc[:,'x1'][y_predict==0],X.loc[:,'x2'][y_predict==0])
    label1 = plt.scatter(X.loc[:,'x1'][y_predict==1],X.loc[:,'x2'][y_predict==1])

    plt.title("predicted data")
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend((label0,label1),('label0','label1'), loc='upper left')
    plt.scatter(centers_i[:,0],centers_i[:,1])
    fig_i.savefig('2d_output/{}.png'.format(i),dpi=500,bbox_inches = 'tight')
```

```python
#生成gif动图
import PIL.Image as Image
def get_gif(pic_dir,n,t=1):
    imgs = []
    for i in range(1,n):
        pic_name = '{}/{}.png'.format(pic_dir,i)
        temp = Image.open(pic_name)
        imgs.append(temp)
    save_name = '{}.gif'.format(pic_dir)
    imgs[0].save(save_name, save_all=True, append_images=imgs, duration=3)
    return save_name


pic_dir = '2d_output'
save_name = get_gif(pic_dir,10,1)
print('制作完成。文件为:{}'.format(save_name))
```
